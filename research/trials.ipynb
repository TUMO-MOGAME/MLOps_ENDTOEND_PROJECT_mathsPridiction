```python
summary_stats = train_data.describe()

# Checking for value counts of categorical columns
gender_counts = train_data['gender'].value_counts()
race_ethnicity_counts = train_data['race_ethnicity'].value_counts()
education_counts = train_data['parental_level_of_education'].value_counts()
lunch_counts = train_data['lunch'].value_counts()
test_preparation_counts = train_data['test_preparation_course'].value_counts()
```

```python
# Check if there are any missing values in the dataset
missing_values = train_data.isnull().sum()
missing_values
```

```python


# If there are missing values, fill them with the median for numerical columns
# and the mode for categorical columns
train_data['math_score'].fillna(train_data['math_score'].median(), inplace=True)
train_data['reading_score'].fillna(train_data['reading_score'].median(), inplace=True)
train_data['writing_score'].fillna(train_data['writing_score'].median(), inplace=True)
train_data['gender'].fillna(train_data['gender'].mode()[0], inplace=True)
train_data['race_ethnicity'].fillna(train_data['race_ethnicity'].mode()[0], inplace=True)
train_data['parental_level_of_education'].fillna(train_data['parental_level_of_education'].mode()[0], inplace=True)
train_data['lunch'].fillna(train_data['lunch'].mode()[0], inplace=True)
train_data['test_preparation_course'].fillna(train_data['test_preparation_course'].mode()[0], inplace=True)
```

```python
# 2. Create dummy variables:
# Convert categorical columns to dummy variables (one-hot encoding)
train_data = pd.get_dummies(train_data, columns=['gender', 'race_ethnicity', 'parental_level_of_education', 'lunch', 'test_preparation_course'])
```
